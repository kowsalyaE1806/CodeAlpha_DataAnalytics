```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
print("Dataset loaded successfully!")
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), 
            annot=True, 
            cmap='coolwarm')
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Survived', y='Age', 
            data=df, 
            palette='Set2')
plt.title('Age Distribution by Survival')
plt.xlabel('0 = Died, 1 = Survived')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Survived', y='Age', 
            data=df, 
            palette='Set2')
plt.title('Age Distribution by Survival')
plt.xlabel('0 = Died, 1 = Survived')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Sex', hue='Survived', 
              data=df, 
              palette='Set1')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(['Died', 'Survived'])
plt.show()
```

```python
plt.figure(figsize=(7,7))
df['Survived'].value_counts().plot(
    kind='pie',
    labels=['Died','Survived'],
    colors=['#ff6b6b','#51cf66'],
    autopct='%1.1f%%',
    startangle=90)
plt.title('Overall Survival Rate')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Fare'], bins=30, 
             color='teal', kde=True)
plt.title('Fare Distribution of Passengers')
plt.xlabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Pclass', hue='Survived',
              data=df,
              palette='muted')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class (1=First, 2=Second, 3=Third)')
plt.ylabel('Count')
plt.legend(['Died','Survived'])
plt.show()
```

```python
print("=== VISUALIZATION CONCLUSIONS ===")
print("1. Only 38.4% of passengers survived")
print("2. Females survived 74% vs Males 19%")
print("3. First class passengers survived more")
print("4. Third class had highest death count")
print("5. Most passengers paid low fares")
print("6. Survivors were slightly younger in age")
```
