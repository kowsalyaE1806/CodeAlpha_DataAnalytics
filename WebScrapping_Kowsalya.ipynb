```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "http://books.toscrape.com/"
response = requests.get(url)

print("Connection Status:", response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
books = soup.find_all('article', class_='product_pod')

print("Total books found:", len(books))
```

```python
all_books = []

for page in range(1, 6):
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    books = soup.find_all('article', class_='product_pod')
    
    for book in books:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        rating = book.p['class'][1]
        
        all_books.append({
            'Title': title,
            'Price': price,
            'Rating': rating
        })

df2 = pd.DataFrame(all_books)
print("Total books scraped:", len(df2))
df2.head(10)
```

```python
df2.to_csv('books_data_100.csv', index=False)

print("=== WEB SCRAPING CONCLUSIONS ===")
print("1. Successfully scraped books.toscrape.com")
print("2. Scraped 5 pages of data")
print("3. Collected 100 books total")
print("4. Extracted Title, Price and Rating")
print("5. Created custom dataset using BeautifulSoup")
print("6. Saved complete data to CSV file")
```
